```python
line1 = ["", "", ""]
line2 = ["", "", ""]
line3 = ["", "", ""]

map = [line1, line2, line3]

print("Hiding your treasure! X marks the spot.")
position = input()

letter = ord(position[0]) - ord("A")
num = int(position[1]) - 1

map[letter][num] = "X" 

print(f"{line1}\n{line2}\n{line3}")
```

```python
total = 0

print("Thank you for choosing PyZa Pizza Deliveries!")

piza_size = input("What size pizza do you want? S, M, or L: ")

if (piza_size.upper() == "S"):
    total += 15
elif (piza_size.upper() == "M"):
    total += 20
elif (piza_size.upper() == "L"):
    total += 25

pepperoni_option = input("Do you want pepperoni? Y or N: ")

if (pepperoni_option.upper() == "Y" and piza_size.upper() == "S"):
    total += 2
elif (pepperoni_option.upper() == "Y" and (piza_size.upper() == "M") or (piza_size.upper() == "L")):
    total += 3

extra_cheese = input("Do you want extra cheese? Y or N: ")

if (extra_cheese.upper() == "Y"):
    total += 1

print(f"Thank you for choosing PyZa Pizza Deliveries!\nYour final bill is: ${total}.")


```
