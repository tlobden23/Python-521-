```python
# 2.27
print("2.27")

# (Quadratic formula) The forumla that calculates roots for a quadratic equation ax**2+bx+c
# is the quadratic formula 
# x = -b+-sqrt(b**2-4ac)/2a
# Because the square root of a negative is imaginary, one can use the expression under the square root (known as the discriminant) to check for the type of root.
# If the discriminant is negative, the roots are imaginary. If the discriminant is zero, there is only one root. If the discriminant is positive, there are two roots.

# a. Write a program that uses the quadratic formula to generate real roots, that is, ignores the imaginary roots.
# Use the discriminant to determine whether there is one root or two roots and then prints the appropriate answer.

import math

def quadratic_formula(a: float, b: float, c:float):
    # get the discriminant
    discriminant = pow(b,2) - 4 * a * c

    # check if discriminant is less than 1
    if discriminant < 0 :
        print("The roots are imaginary.")

    # check if discriminant is equal to 0 meaning 1 root
    elif discriminant == 0:
        root = -b / (2*a)
        print(f"There is only one root. {root}")

    # else statement if none of above statements match 2 roots
    else: 
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        print(f"The two roots are {root1} and {root2}")

# b. Python uses the letter 'j' to represent the mathematical imaginary number 'i' (a convention used in electrical engineering).
# However, the Python ‘j’ must always be preceeded by a number. That is, '1j' is eqivalent to the mathematical 'i'. Add the ability to handle imaginary roots to your program.

def quadratic_formula_imaginary(a: float, b: float, c: float):
    # get the discriminant
    discriminant = pow(b,2) - 4 * a * c

    # check if discriminant is less than 1
    if discriminant < 0 :
        real_part = -b / (2*a)
        imagine_part = math.sqrt(-discriminant) / (2*a)
        # add j because it's imaginary root
        root1 = f"{real_part}+{imagine_part}j"
        root2 = f"{real_part}-{imagine_part}j"
        print(f"The two imaginary roots are {root1} and {root2}")
    
    # check if discriminant is equal to 0 meaning 1 root
    elif discriminant == 0:
        root = -b / (2*a)
        print(f"There is only one root. {root}")
    else: 
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        print(f"The two roots are {root1} and {root2}")
```

```python
# Leap Year Problem

# Exercise 1
def leap_year_boolean(year:int):

    # check if 400 is divisible
    if year % 400 == 0:
        return True
    
    # check if 100 is divisible, but not 400
    elif (year % 100 == 0) and (year % 400 != 0):
        return False
    
    # check if 4 is divisible, but not 100
    elif (year % 4 == 0) and (year % 100 != 0):
        return True
    
    # if none of the above statement matches then False
    else:
        return False

print("Exercise 1")
print(leap_year_boolean(2000))
print(leap_year_boolean(1900))
print(leap_year_boolean(2024))
print(leap_year_boolean(2023))
print()

# Exercise 2
def leap_year_boolean_chained(year: int):
    return (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0)

print("Exercise 2")
print(leap_year_boolean_chained(2000))
print(leap_year_boolean_chained(1900))
print(leap_year_boolean_chained(2024))
print(leap_year_boolean_chained(2023))
```

```python
# 2.22 
print(2.22)

# Write a program that checks to see if a number N is prime. 
# A simple approach checks all numbers from 2 up to N, but after some point numbers are checked that need not be checked. 
# For example, numbers greater than need not be checked. Write a program that checks for primality and avoids those unnecessary checks. 
# Remember to import the math module.

def is_prime(num: int):
    # check if the parameter is less than 2
    if num < 2:
        return False
    
    # calculate the limit which is square root of num
    limit = int(math.sqrt(num))

    # check for each num up to limit
    for i in range(2, limit + 1):

        # check if it's divisible, 
        if num % i == 0:
            return False
    
    # return True because no number less than limit is divisible
    return True

print(
    "The reason we only need to check up to the square root of N is that after the square root,"
    "we would be checking factors that correspond to numbers we have already tested.\n"
    "For example, if we are checking 100, the square root of 100 is 10.\n"
    "After 10, dividing 100 by 11 gives a number less than 10.\n"
    "This shows that any factor greater than the limit will pair with a factor smaller than the limit, "
    "which we have already checked."
)
```

### 4. Nested for loops

```python
# Assume that the two matrices are stored as nested lists. For example, with
# A = [[1,2,0],[-1,2,1]], and
# B = [[1,2],[1,0],[-1,-2]],
# their product C = AB = [[3,2],[0,-4]]

def matrix_multiply(A, B):
    # number of rows of A
    rows_A = len(A)

    # number of columns in A and B
    columns_A = len(A[0])
    columns_B = len(B[0])

    # create empty_matrix
    empty_matrix = []

    # iterate over each row in A
    for i in range(rows_A):
        row = []
        # iterate over each column in B
        for j in range(columns_B):
            total = 0
            # iterate over A's row and B's column
            for k in range(columns_A):
                total += A[i][k] * B[k][j]
            # add value to current row
            row.append(total)  
        # add completed row to matrix        
        empty_matrix.append(row)       

    return empty_matrix


A = [[1, 2, 0], [-1, 2, 1]]
B = [[1, 2], [1, 0], [-1, -2]]

C = matrix_multiply(A, B)
print(C)
```
