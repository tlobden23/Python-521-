### 2. Sets

```python
# a. Given the sets
# s1 = {'a', 'e', 'i', 'o', 'u'}
# s2 = {'a', 'e', 'i'}
# s3 = {'a', 'e', 'f', 'o'}
# how will you use built-in set methods to find if s2 and s3 are subsets of s1?

s1 = {'a', 'e', 'i', 'o', 'u'}
s2 = {'a', 'e', 'i'}
s3 = {'a', 'e', 'f', 'o'}

print(f"a. Is s2 a subset of s1? {s2.issubset(s1)}")
print(f"Is s3 a subset of s1? {s3.issubset(s1)}")
print(f"Is s2 and s3 subset of s1? {(s2.issubset(s1) and (s3.issubset(s1)))}")

# b. Given two sets A and B, the Symmetric Difference between two is defined as A triangle B = (A U B) / (A inverse U B).
# Give two different methods by which you will find the symmetric difference between s1 and s3 above. 
# What built-in functions will you use for each method?

print(f"b. Using built-in function = {s1.symmetric_difference(s3)}")
print(f"Not using built-in function = {s1.union(s3).difference(s1.intersection(s3))}")
```

### 3. Flower garden

```python
import math

print("Calculate Garden requirements")
print("-----------------------------")

# user inputs
side_garden = float(input("Enter length of side of garden (feet): "))
depth_garden = float(input("Enter depth of garden soil (feet): "))
depth_filler = float(input("Enter depth of filler (feet): "))

# variables
radius = side_garden / 4
area_circle = math.pi * pow(radius, 2)
area_one_semicircle = 0.5 * math.pi * pow(radius, 2)
area_soil = area_circle + 4 * area_one_semicircle
area_square = pow(side_garden, 2)
area_filler = area_square - area_soil

# volumes in cubic yards
soil_each_semicircle = (area_one_semicircle * depth_garden) / 27
soil_circle = (area_circle * depth_garden) / 27
total_soil = (area_soil * depth_garden) / 27
total_filler = (area_filler * depth_filler) / 27


print("-----------------------------")
print("Requirements\n")

print(f"Soil for each semicircle garden: {round(soil_each_semicircle,1)} cubic yards")
print(f"Soil for the circle garden: {round(soil_circle,1)} cubic yards")
print(f"Total soil for the garden: {round(total_soil, 1)} cubic yards")
print(f"Total filler for the garden: {round(total_filler, 1)} cubic yards")
```
