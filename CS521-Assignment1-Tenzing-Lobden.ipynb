```python
# a. Given the string "Monty Python"
string = "Monty Python"

# 1. Write an expression to print the first character.
print(f"1. {string[0]} or {string[-12]}")

# 2. Write an expression to print the last character.
print(f"2. {string[11]} or {string[-1]}")

# 3. Write an expression including len() to print the last character.
print(f"3. length = {len(string)} and {string[len(string) - 1]}")

# 4. Write an expression that prints "Monty".
print(f"4. {string[:5]} or {string[:-7]}")
```
